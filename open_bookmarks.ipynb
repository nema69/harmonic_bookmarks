```python
import requests

file = open("./bookmarks.txt")
file_content = file.read()
```

```python
def teile_string_auf(input_string):
    teile = []
    aktueller_string = ""
    suche_q = True  # Steuerungsvariable, die angibt, ob nach 'q' oder nach '-' gesucht wird

    for zeichen in input_string:
        if suche_q and zeichen == 'q':
            teile.append(aktueller_string)
            aktueller_string = ""  # Zurücksetzen des aktuellen Strings für den nächsten Teil
            suche_q = False  # Jetzt suchen wir nach '-'
        elif not suche_q and zeichen == '-':
            aktueller_string = ""  # Zurücksetzen des aktuellen Strings für den nächsten Teil
            suche_q = True  # Zurück zur Suche nach 'q'
        else:
            aktueller_string += zeichen

    # if aktueller_string:  # Den Rest des Strings hinzufügen, wenn er nicht leer ist
    #     teile.append(aktueller_string)

    return teile
```

```python
strings = teile_string_auf(file_content)
print(strings)
```

```python

file_path = './extracted_bookmarks.txt'
for id in strings:
    url = "https://hacker-news.firebaseio.com/v0/item/" + id +".json"
    response = requests.get(url)
    if response.status_code == 200:
        data_dict = response.json()
        #print(data_dict)
    # Sie können auf Daten im Dictionary zugreifen wie auf jedes andere Dictionary auch
        value = data_dict.get('url', 'NA')
        with open(file_path, 'a') as file:
            if value != "NA":
                file.write(value+ "\n")
            else:
                file.write("https://news.ycombinator.com/item?id=" + id+ "\n")
    else:
        print("Error while fetching data:", response.status_code)
```
